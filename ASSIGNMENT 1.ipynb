```python
#3. Calculate the mean of each column of the dataset.

#means = iris_data.mean()
means = iris_data.mean(numeric_only=True)
print("\nMean of each column:")
print(means)
```

```python
#4. Check for the null values present in the dataset.

null_values = iris_data.isnull().sum()
print("\nNull values in the dataset:")
print(null_values)
```

```python
#5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

# Visualization 1: Boxplot for Sepal Length by Classification
plt.figure(figsize=(10, 6))
sns.boxplot(x='Classification', y='SL', data=iris_data)
plt.title('Boxplot of Sepal Length by Classification')
plt.show()

```

```python
# Visualization 2: Scatter plot for Sepal Length and Width
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SL', y='SW', hue='Classification', data=iris_data)
plt.title('Scatter Plot for Sepal Length and Width')
plt.show()
```

```python
# Visualization 3: Bar chart for Sepal Width (SW)
plt.figure(figsize=(10, 6))
sns.barplot(x=iris_data['Classification'], y='SW', data=iris_data)
plt.title('Bar Chart for Sepal Width by Species')
plt.show()
```

```python
# Visualization 4: Bar chart for Petal Width (PW)
plt.figure(figsize=(10, 6))
sns.barplot(x=iris_data['Classification'], y='PW', data=iris_data)
plt.title('Bar Chart for Petal Width by Species')
plt.show()

#This bar chart represents the average (mean) petal width for each species in the Iris dataset. Each bar corresponds to a different species, and the height of the bar indicates the average petal width for that particular species.
```
